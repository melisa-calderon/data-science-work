```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/datasets/five-thirty-eight-datasets/master/datasets/alcohol-consumption/data/drinks.csv')
```

```python
data[['wine_servings']].plot.hist(bins=[0,50,100,150,200,250,300,350],rwidth=0.8)
plt.title('Frequency of the Average Wine Servings in Different Countries')
plt.xlabel('Average Serving of Wine')
```
