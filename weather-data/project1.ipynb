```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('weather_date.csv')
```

```python
act_mean_temp = data[['actual_mean_temp']]
```

```python
act_min_temp = data[['actual_min_temp']]
```

```python
dates = data[['date']]
```

```python
act_max_temp = data[['actual_max_temp']]
```

```python
weather_data = pd.concat([act_mean_temp,act_min_temp, act_max_temp])
```

```python
weather_data.plot()
plt.xlabel('Days After July 1st 2014')
plt.ylabel('Temperature')
plt.title('Average Temperature in the US:From July 1st, 2014 to June 30th, 2015')
```
